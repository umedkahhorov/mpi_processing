## üî∫ Triangle Smoothing ‚Äì  Overview

**Triangle smoothing** is a weighted moving average where weights **taper off symmetrically like a triangle**:

* The center point gets the most weight.
* Neighbors get less weight as you move away.

This makes the smoothing gentler and better at reducing noise than a simple box average.

---

**Frequency Response (how it filters different frequencies):**

$$
T(R,\omega)
= \frac{1}{R^2}
  \,\frac{\sin^2\!\bigl(\tfrac{R\,\omega\,\Delta t}{2}\bigr)}
         {\sin^2\!\bigl(\tfrac{\omega\,\Delta t}{2}\bigr)}
$$

**To implement in the frequency domain:**

* **Step 1:** Compute the **2D Fourier Transform** of your data array.
* **Step 2:** Compute the frequency response $T(R,\omega)$ on the same frequency grid.
* **Step 3:** Multiply the transformed data by this frequency response (this is the convolution).
* **Step 4:** Compute the **inverse Fourier Transform** to return to the time-space domain.

This approach directly applies the filter in the frequency domain and is often more efficient for very large windows (especially when the smoothing radius $R$ is large).

---

### **Interpretation & How to Implement (Pseudo-Code)**

```
# Triangle smoothing pseudo-code

# Inputs:
#   data: 2D array [space, time]
#   R_time: smoothing radius along time axis
#   R_space: smoothing radius along space axis

# 1. Box smoothing along time (1D moving average)
data = box_smooth(data, radius=R_time, axis='time')

# 2. Box smoothing along time again to make triangle smoothing
data = box_smooth(data, radius=R_time, axis='time')

# 3. Box smoothing along space
data = box_smooth(data, radius=R_space, axis='space')

# 4. Box smoothing along space again to make triangle smoothing
data = box_smooth(data, radius=R_space, axis='space')

# Optional: Fold edges to keep constant signals unchanged
data = fold_edges(data)
```

**Notes:**

* *Triangle smoothing is simply applying box smoothing twice.*
* *This yields a symmetric, low-pass filter that preserves low frequencies.*
* *Folding edges avoids distortion near boundaries.*
* *Recursive algorithms can be used internally to speed up long windows.*

````python
def seismic_slice_extraction(segyfile,df,col_idx,value,ext=True):
    """
    extract a seismic slice from a segy file based on a inline or crossline value
    :param segyfile: str, path to the segy file
    :param df: pandas DataFrame, DataFrame containing inline/crossline headers
    :param col_idx: int, index of the column in df to search for the value
    :param value: int, the inline or crossline value to match
    :param ext: bool, if True, return the extent of the slice, otherwise return only the slice as a numpy array
    """
    # select the column by index
    col_series = df.iloc[:, col_idx]
    # inline/xline extent
    min_, max_ = df.iloc[:, df.columns != df.columns[col_idx]].squeeze().agg(['min','max'])

    # mask for exact match
    mask = col_series == value
    if mask.any():
        idxs = df.index[mask].tolist()
        #print(f"Row min/max indexes: ",idxs[0],idxs[-1])
    if not mask.any():
        print(f"Slice {value} not found in column #{col_idx}.")

    with segyio.open(filename,'r',ignore_geometry=True) as f:
        f.mmap()
        n_twt    = f.samples
        selected = np.vstack([f.trace.raw[i] for i in idxs])
    extent = [min_,max_,n_twt[0],n_twt[-1]]
    if ext:
        return selected,extent
    else:
        return selected
````

````python
def mean(x,w):
    m = np.cumsum(x)
    m[w:] = m[w:] - m[:-w]
    m /= w
    return m

def box_smoothing_1d(x, N):
    """
    1D box smoothing with folding edges
    """
    n = len(x)
    # Pad by folding edges
    leftx = x[N-1::-1]
    rightx = x[:-N-1:-1]
    padded = np.concatenate((leftx, x, rightx))
    # Moving average
    smoothed = mean(padded, N)
    # Extract part matching original length
    out = smoothed[N:N+n]
    return out
def triangle_smooth_1d(x, N):
    """

    """
    y1 = box_smoothing_1d(x, N)
    y2 = box_smoothing_1d(y1, N)
    return y2
def triangle_smooth_2d(data, N_time, N_trace):
    """
    2D box smoothing with folding edges
    data: 2D array [space, time]
    N_time: window size in time samples
    N_trace: window size in space samples
    """
    smoothed = np.zeros_like(data)
    # smooth along time axis
    for trace in range(data.shape[0]):
        smoothed[trace,:] = triangle_smooth_1d(data[trace,:], N_time)
    # smooth along space axis
    for t in range(smoothed.shape[1]):
        smoothed[:,t] = triangle_smooth_1d(smoothed[:,t], N_trace)
    return smoothed
````

````python
def triangle_smoothing(data: np.ndarray, rect, diff=None, box=None, nrep=1, adj=False):
    """
    Multi-dimensional triangle/box smoothing (like sfsmooth).
    
    Parameters:
        data : Input array (1D, 2D, or 3D).
        rect : List of smoothing radii per axis.
        diff : List of booleans: derivative per axis.
        box  : List of booleans: box filter (True) or triangle (False).
        nrep : Number of repetitions.
        adj  : Use adjoint operator.
    Returns:
        Smoothed array.
    """
    ndim = data.ndim

    # Default to no differentiation and triangle smoothing
    if diff is None:
        diff = [False] * ndim  # Default: blur all dimensions
    if box is None:
        box = [False] * ndim # Default: use triangle blur

    # Copy input so we don't modify original
    result = data.copy()

    # Repeat smoothing as many times as requested (e.g., to approximate Gaussian)
    for _ in range(nrep):
        # Loop over each axis separately
        for axis in range(ndim):
            if rect[axis] <= 1:
                # Skip trivial smoothing
                continue
            # Apply smoothing on this axis
            result = _smooth_axis(
                result, axis, rect[axis], diff[axis], box[axis], adj
            )
    return result


def _smooth_axis(data, axis, nbox, der, is_box, adj):
    """
    Apply smoothing along a single axis using integration and folding,
    following the same algorithm as sf_smooth2 in C.
    """
    nx = data.shape[axis]     # Length along this dimension
    npad = 2 * nbox           # total padding on both sides
    np_tot = nx + npad        # total length including padding

    # Create an output array with same shape
    smoothed = np.zeros_like(data, dtype=np.float32)

    # Iterate over all slices along the axis (other dimensions held fixed)
    it = np.nditer(
        np.zeros(data.shape[:axis] + data.shape[axis+1:]),
        flags=['multi_index']
    )
    while not it.finished:
        # Build an index tuple to extract the 1D line along the axis
        # Grab one line of data (like one row of pixels)
        idx = list(it.multi_index)
        idx.insert(axis, slice(None))
        x = data[tuple(idx)].astype(np.float32)

        # Temporary array for cumulative sums and shifted weights
        tmp = np.zeros(np_tot, dtype=np.float32)

        # Compute normalization weight
        # For a box: 1/(2N-1), for triangle: 1/N^2
        if is_box:
            wt = 1.0 / (2 * nbox - 1)
        else:
            wt = 1.0 / (nbox * nbox)

        # === triple2 step ===
        # Build shifted weighted sums of the signal
        # This is equivalent to finite differences in C code:
        #   tmp[1:1+nx] = +wt * x
        #   tmp[2*nbox:2*nbox+nx] = -wt * x
        # or for triangle:
        #   tmp = -wt * x
        #   tmp shifted by N = +2wt * x
        #   tmp shifted by 2N = -wt * x
        if is_box:
            tmp[1:1+nx] += +wt * x
            tmp[2*nbox:2*nbox+nx] += -wt * x
        else:
            tmp[0:nx] += -wt * x
            tmp[nbox:nbox+nx] += +2 * wt * x
            tmp[2*nbox:2*nbox+nx] += -wt * x

        # === doubint2 step ===
        # Perform forward cumulative sum (integration)
        t = 0.0
        for i in range(np_tot):
            t += tmp[i]
            tmp[i] = t
        # For triangle smoothing without differentiation,
        # perform backward cumulative sum (double integration)
        if not is_box and not der:
            t = 0.0
            for i in reversed(range(np_tot)):
                t += tmp[i]
                tmp[i] = t

        # === fold2 step ===
        # Fold mirrored contributions back into main signal
        # This preserves constant signals and avoids edge artifacts
        y = np.copy(x)

        # Middle part (no folding needed)
        y[:] = tmp[nbox:nbox+nx]

        # Right reflections: mirror and sum back
        j = nbox + nx
        while j < np_tot:
            for i in range(min(nx, np_tot - j)):
                y[nx - 1 - i] += tmp[j + i]
            j += nx
            for i in range(min(nx, np_tot - j)):
                y[i] += tmp[j + i]
            j += nx

        # Left reflections: mirror and sum back
        j = nbox
        while j >= 0:
            for i in range(min(nx, j)):
                y[i] += tmp[j - 1 - i]
            j -= nx
            for i in range(min(nx, j)):
                y[nx - 1 - i] += tmp[j - 1 - i]
            j -= nx

        # Store smoothed output in correct slice
        smoothed[tuple(idx)] = y

        it.iternext()

    return smoothed
````

````python
"""
    Multi-dimensional triangle/box smoothing, equivalent to sfsmooth from Madagascar.

    This smoothing replaces each sample with a weighted average of its neighbors.
    For triangle smoothing, the weights taper linearly from the center outward,
    and can be interpreted as the cross-correlation of two box filters:

        T(Z) = B(Z) * B(Z^{-1})

    where:
        B(Z) = (1 - Z^N) / [N*(1 - Z)]

    Triangle smoothing approximates Gaussian smoothing when repeated multiple times,
    and it efficiently attenuates high frequencies while preserving trends.

    Parameters:
        data : ndarray
            Input array (1D, 2D, or 3D) to be smoothed.
        
        rect : list of int
            Smoothing radius for each axis.
            For axis i, rect[i]=N means the triangle filter spans 2N samples.
            Larger radius results in stronger smoothing and more low-pass behavior.
            For example, rect=[12,6] smooths with a 12-sample radius along the first axis
            and a 6-sample radius along the second axis.
        
        diff : list of bool, optional
            If True on an axis, computes the *derivative* instead of smoothing along that axis.
            This corresponds to the three-point filtering step in Claerbout (1992):
            
                y_t = 2 x_t - x_{t-N} - x_{t+N}
        
            which emphasizes edges and suppresses constant components.
            Defaults to False for all axes (i.e., pure smoothing).
        
        box : list of bool, optional
            If True on an axis, uses a rectangular (boxcar) filter instead of triangle.
            A box filter is a simple moving average:
            
                B(Z) = (1 - Z^N) / [N*(1 - Z)]
            
            The triangle filter is the convolution of a box filter with itself,
            and has a smoother, more Gaussian-like shape.
            Defaults to False for all axes (i.e., triangle smoothing).
        
        nrep : int, optional
            Number of times to repeat smoothing.
            Repeating the smoothing multiple times further approximates a Gaussian
            (per the central limit theorem), producing stronger noise suppression.
            For example, nrep=3 applies the smoothing filter three times.
            Defaults to 1.
        
        adj : bool, optional
            If True, applies the adjoint (transpose) of the smoothing operator.
            This uses the anti-causal integration (backward recursion), which can be
            important in least-squares inversion workflows.
            Defaults to False (standard forward smoothing).
    
    Returns:
        smoothed : ndarray
            Smoothed array with the same shape as the input.

    Notes:
        - Smoothing is implemented via efficient cumulative sums ("integration")
          and edge folding, avoiding explicit convolution over long windows.
        - Folding mirrored edges ensures that constant signals remain unchanged
          and that the filter is truly low-pass (passing frequency œâ=0).
        - This is equivalent to applying causal and anti-causal integration in sequence,
          matching the formulation in Claerbout (1992).

    References:
        Claerbout, J. F. (1992). Earth Soundings Analysis: Processing versus Inversion.
        Fomel, S. (2016). Notes on triangle smoothing.
    """
````
